### 1. Пространство выборочного множества
- Обозначается как Ω (омега); точки обозначаются как ω (омега).

### 2. Вероятность события
- Вероятность события находится в диапазоне от 0 до 1.
  - Если вероятность равна 1, событие достоверное.
  - Если вероятность равна 0, событие недостоверное.
  - Если вероятность находится в диапазоне от 0 до 1, событие случайное.

### 3. Размер множества
- Пустое множество имеет размер 0.

### 4. Подмножество
- A содержится в B (A ⊆ B), если все элементы A находятся в B.
- B содержит A (B ⊇ A), если все элементы A находятся в B.

### 5. Операции над множествами
#### 5.1. Дополнение
- Дополнение множества A — это множество всех элементов, которые не принадлежат A (Aᶜ).

#### 5.2. Объединение (ИЛИ)
- Объединение двух множеств A и B (A ∪ B) — это множество всех элементов, которые принадлежат хотя бы одному из них.

#### 5.3. Пересечение (И)
- Пересечение A и B (A ∩ B) — это множество, состоящее из всех элементов, которые принадлежат одновременно обоим множествам.

#### 5.4. Разность
- Разность множества A и B (A \ B) — это множество всех элементов, которые принадлежат A, но не принадлежат B.
- Разность не является ассоциативной.

#### 5.5. Симметрическая разность
- Симметрическая разность (A △ B) — это множество, состоящее из элементов, которые принадлежат только одному из множеств.

### 6. Коммуникативный закон
- Объединение и пересечение множеств являются коммутативными операциями:
  - A ∪ B = B ∪ A
  - A ∩ B = B ∩ A

### 7. Ассоциативный закон
- Объединение и пересечение множеств являются ассоциативными операциями:
  - (A ∪ B) ∪ C = A ∪ (B ∪ C)
  - (A ∩ B) ∩ C = A ∩ (B ∩ C)

### 8. Дистрибутивный закон
- Объединение и пересечение множеств подчиняются дистрибутивному закону:
  - A ∪ (B ∩ C) = (A ∪ B) ∩ (A ∪ C)
  - A ∩ (B ∪ C) = (A ∩ B) ∪ (A ∩ C)

### 9. Законы де Моргана
1. Закон 1: Дополнение объединения равно пересечению дополнений.
   - (A ∪ B)ᶜ = Aᶜ ∩ Bᶜ
2. Закон 2: Дополнение пересечения равно объединению дополнений.
   - (A ∩ B)ᶜ = Aᶜ ∪ Bᶜ

### 10. Непересекающиеся множества
- Непересекающиеся множества имеют пустое пересечение (A ∩ B = ∅).
