## Классическое определение вероятности:

Вероятность наступления события A в некотором испытании равна отношению m/n, где:
- n – общее число всех равновозможных, элементарных исходов данного испытания, которые образуют полную группу событий;
- m – количество элементарных исходов, благоприятствующих событию A.

## Задача 1

В урне находится 15 белых, 5 красных и 10 чёрных шаров. Наугад извлекается 1 шар, найти вероятность того, что он будет:
- а) белым
- б) красным
- в) чёрным

### Решение:

Всего в урне: 15 + 5 + 10 = 30 шаров

По классическому определению:

а) P(белый) = 15/30 = 1/2 = 0.5

б) P(красный) = 5/30 = 1/6 ≈ 0.167

в) P(черный) = 10/30 = 1/3 ≈ 0.333

Проверка: P(белый) + P(красный) + P(черный) = 0.5 + 0.167 + 0.333 = 1

### Ответ: 1/2; 1/6; 1/3

## Задача 2

В магазин поступило 30 холодильников, пять из которых имеют заводской дефект. Случайным образом выбирают один холодильник. Какова вероятность того, что он будет без дефекта?

### Решение:

30 - 5 = 25 холодильников не имеют дефекта.

По классическому определению:
P(без дефекта) = 25/30 = 5/6 ≈ 0.833

### Ответ: 5/6

## Задача 5

Найти вероятность того, что при бросании двух игральных костей в сумме выпадет:
- а) пять очков
- б) не более четырёх очков
- в) от 3 до 9 очков включительно

### Решение:

Всего исходов: 6 × 6 = 36 

а) Благоприятствующие комбинации для суммы 5:
(1,4), (2,3), (3,2), (4,1)
m = 4
P(5) = 4/36 = 1/9 ≈ 0.111

б) Благоприятствующие комбинации для суммы ≤ 4:
2 очка: (1,1)
3 очка: (1,2), (2,1)
4 очка: (1,3), (2,2), (3,1)
m = 6
P(≤4) = 6/36 = 1/6 ≈ 0.167

в) Противоположное событие (сумма < 3 или > 9):
2 очка: (1,1)
10 очков: (4,6), (5,5), (6,4)
11 очков: (5,6), (6,5)
12 очков: (6,6)
m = 7
P(противоположное) = 7/36
P(3-9) = 1 - 7/36 = 29/36 ≈ 0.806

### Ответ: 1/9; 1/6; 29/36

## Задача 6

Найти вероятность того, что при броске двух игральных костей произведение очков:
- а) будет равно семи
- б) окажется не менее 20
- в) будет чётным

### Решение:

Всего исходов: 6 × 6 = 36

а) Число 7 нельзя получить как произведение двух целых чисел от 1 до 6
P(7) = 0

б) Произведение ≥ 20:
(4,5), (5,4), (4,6), (6,4), (5,5), (5,6), (6,5), (6,6)
m = 8
P(≥20) = 8/36 = 2/9 ≈ 0.222

в) Нечетные произведения:
(1,1), (1,3), (1,5), (3,1), (3,3), (3,5), (5,1), (5,3), (5,5)
m = 9
P(четное) = 1 - 9/36 = 27/36 = 3/4 = 0.75

### Ответ: 0; 2/9; 3/4

## Задача 7

В лифт 20-этажного дома на первом этаже зашли 3 человека. Найти вероятность того, что:
- а) они выйдут на разных этажах
- б) двое выйдут на одном этаже
- в) все выйдут на одном этаже

### Решение:

Всего исходов: 19³ = 6859 (19 этажей, не считая первый, для каждого из трех пассажиров)

а) Размещение по разным этажам:
P(разные) = 19 × 18 × 17 / 6859 = 5814/6859 ≈ 0.848

в) Все на одном этаже:
P(один) = 19/6859 ≈ 0.003

б) Двое на одном этаже:
P(два) = 1 - P(разные) - P(один) = 1026/6859 ≈ 0.149

### Ответ: 5814/6859; 1026/6859; 19/6859

## Задача 8

Подбрасывается 10 монет. Найти вероятность того, что:
- а) на всех монетах выпадет орёл
- б) на 9 монетах выпадет орёл, а на одной – решка
- в) орёл выпадет на половине монет

### Решение:

Всего исходов: 2¹⁰ = 1024

а) Для всех орлов только один благоприятный исход
P(все орлы) = 1/1024

б) Выбрать 1 монету из 10 для решки: C(10,1) = 10
P(9 орлов) = 10/1024 = 5/512

в) Выбрать 5 монет из 10 для орлов: C(10,5) = 252
P(5 орлов) = 252/1024 = 63/256

### Ответ: 1/1024; 5/512; 63/256

## Задача 9

На семиместную скамейку случайным образом рассаживается 7 человек. Какова вероятность того, что два определённых человека окажутся рядом?

### Решение:

Всего исходов: P(7) = 7! = 5040 (перестановки 7 человек)

Для каждой пары соседних мест (6 пар) два определенных человека могут сесть 2! способами
Остальные 5 человек могут сесть 5! способами

Благоприятные исходы: 6 × 2! × 5! = 1440

P(рядом) = 1440/5040 = 2/7 ≈ 0.286

### Ответ: 2/7

## Задача 10

На шахматную доску из 64 клеток ставят наудачу две ладьи, белого и чёрного цвета. С какой вероятностью они не будут «бить» друг друга?

### Решение:

Всего исходов: 64 × 63 = 4032 (первая ладья на любую клетку, вторая - на оставшиеся)

На горизонтали:
- 8 горизонталей
- На каждой горизонтали: 8 × 7 × 2 = 112 способов расстановки (2 - перестановка ладей)
- Всего на горизонталях: 8 × 112 = 448 способов

На вертикали: аналогично 448 способов

Всего неблагоприятных исходов: 896 способов

P(не бьют) = (4032 - 896)/4032 = 3136/4032 = 14/18 ≈ 0.778

### Ответ: 14/18

## Задача 11

Какова вероятность того, что в четырех сданных картах будет один туз и один король?

### Решение:

В колоде 36 карт.

Общее число исходов: C(36,4) = 58905 (способов выбрать 4 карты из 36)

Благоприятные исходы:
- C(4,1) способов выбрать одного туза
- C(4,1) способов выбрать одного короля
- C(28,2) способов выбрать две другие карты

Всего благоприятных исходов: C(4,1) × C(4,1) × C(28,2) = 4 × 4 × 378 = 6048

P(туз и король) = 6048/58905 ≈ 0.103

### Ответ: 6048/58905
