# Основы теории вероятностей

Источник: [mathprofi.ru/teorija_verojatnostei.html](http://mathprofi.ru/teorija_verojatnostei.html)

## Что изучает теория вероятностей?

Теория вероятностей изучает вероятностные закономерности массовых однородных случайных событий. Она не пытается предсказать единичный результат, а изучает закономерности при многократном повторении событий.

## Основные термины и понятия

### События и их виды

1. **Достоверное событие** - обязательно произойдет в результате испытания
   - Пример: подброшенная монета упадет вниз

2. **Невозможное событие** - заведомо не произойдет в результате испытания
   - Пример: подброшенная монета улетит вверх

3. **Случайное событие** - может как произойти, так и не произойти
   - Критерий: является следствием случайных факторов, которые нельзя предугадать
   - Пример: выпадение "орла" при броске монеты

### Исходы и равновозможность

- **Исход** - любой результат испытания
- **Равновозможные события** - события, когда ни одно не является более возможным, чем другие
  - Пример: выпадение орла или решки при броске правильной монеты

### Совместность событий

1. **Несовместные события** - появление одного исключает появление других
   - Пример: выпадение орла и решки в одном броске

2. **Совместные события** - появление одного не исключает появление другого
   - Пример: дождь и гроза одновременно

### Противоположные события
- Обозначаются той же буквой с чертой сверху
- Всегда образуют полную группу событий
- Пример: A - выпадение "5" на кубике, Ā - выпадение любого другого числа

### Полная группа событий
- Множество несовместных событий, одно из которых обязательно произойдет
- Пример: все возможные исходы броска игральной кости (1-6)

## Алгебра событий

### Важнейшее правило:
- Сложение событий = логическая связка "ИЛИ"
- Умножение событий = логическая связка "И"

1. **Сумма событий (A + B)**: 
   - Наступит A ИЛИ B ИЛИ оба события
   - Для несовместных - только одно из событий

2. **Произведение событий (A × B)**:
   - Наступят И событие A, И событие B одновременно

## Вероятность события

### Обозначения:
- P(A) или pₐ - вероятность события A

### Классическое определение вероятности:
P(A) = m/n, где:
- n - общее число равновозможных элементарных исходов
- m - число благоприятствующих событию A исходов

### Свойства вероятности:
1. 0 ≤ P(A) ≤ 1
2. P(невозможного события) = 0
3. P(достоверного события) = 1
4. Сумма вероятностей событий полной группы = 1

### Важные принципы:
1. В единичном испытании маловероятное событие не произойдет
2. Событие с вероятностью близкой к 1 практически достоверно произойдет

### Теорема о сумме вероятностей:
- Для противоположных событий: P(A) + P(Ā) = 1
- Позволяет легко находить вероятность противоположного события: P(Ā) = 1 - P(A)
